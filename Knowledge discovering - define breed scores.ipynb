I use the ranking of the most popular breeds from this website:
http://www.akc.org/news/the-most-popular-dog-breeds-in-america/

```python
# dogs
import csv
import pickle
path = 'mostPopularBreeds.txt'
with open(path) as f:
    reader = csv.reader(f, delimiter="\t")
    d = list(reader)

breed2score = dict()
for row in d:
    try:
        breed2score[  row[0].lower()] = float(int(row[1])+int(row[2]) )/ 2
    except:
        print row

maxx = max( breed2score.values() )
for k in breed2score:
    breed2score[k] = maxx - breed2score[k]
```

```python
for x in sorted( breed2score.items(), key = lambda x: x[1]):
    print x
```
